#### Well if someone told you that -> If you take a population(say 1000000), then if you take repeated samples of the same size from the population, say 1000 sets -> each of size 50(total 50000 samples) and you calculate the mean of those samples. Then if you plot those sample means -> 

### **You will end up with a Normal Distribution, Every Single Time!**

#### What is even more interesting that,***Distribution of the population doesn't matter***. For example, if you have Log normal distribution, or uniform distribution, or any other distribution, If you follow the above steps you will end up with a Normal Distribution!

#### But there are some Constraints -> 

* Independent and Identically Distributed Samples.
* Sample Size should be atleast >30.

## Enough Theory, Let's Dive into Practicality!

#### I will be taking 2 different distributions, and proving that distribution of the population doesn't matter when you apply **Central Limit Theorem (CLT)**

<h1><p style="text-align:center;"> Log Normal Distribution </p> </h1>

### Let's get a brief intro of Log Normal Distribution :

* Definition -> As the name suggests, if you take Natural log of a random variable following a normal distribution you get a log normal distribution.

* It is typically skewed to the right, as you can see in the below plot. Example -> Stock prices, or Income(it is also typically right skewed as we have very few people having high income and huge number of people having low or median income)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define the parameters of the distribution
mu = 1.5  # mean of the logarithm of the random variable
sigma = 0.7  # standard deviation of the logarithm of the random variable
size = 1000000  # number of samples to generate

# Generate samples from the log normal distribution
samples = np.random.lognormal(mean=mu, sigma=sigma, size=size)

# Set the style and color palette
sns.set_style('whitegrid')
sns.set_palette('Dark2')

# Plot the distribution using Seaborn
ax = sns.histplot(samples, kde=True, alpha=0.7, bins=50)

# Customize the plot
ax.set(xlabel='Value', ylabel='Density', title='Log Normal Distribution')
sns.despine(left=True)

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

<h1><p style="text-align:center;"> ü§©Witness the Magic of Central Limit Theoremü§© </p></h1>

#### In the below plot, you can witness how Central Limit Theorem actually works, As you can see that we have proved that the distribution of the samples doesn't matter, Sampling distribution of sample mean will always be Normally distributed.

```python
# Generate 1000 sets of 50 random samples from the distribution
means = np.zeros(1000)  # create a NumPy array to store the mean of each set
for i in range(1000):
    sample_set = np.random.choice(samples, size=50, replace=False) 
    mean = np.mean(sample_set)
    means[i] = mean #appending the mean of each set in the numpy array

# Plot a histogram of the means
sns.histplot(means, kde=True)

# Set the x-label, y-label, and title
plt.xlabel('Mean Value')
plt.ylabel('Density')
plt.title('Histogram of Means')

# Show the plot
plt.show()
```

### Then let me show you using a different distribution -> 

<h1><p style="text-align:center;"> Uniform Distribution </p> </h1>

### Let's get a brief intro of Uniform Distribution :

* Definition -> In technical terms, if a random variable takes on equally likely values within a specific range then it follows a Uniform Distribution. 

* Now in simplistic terms or example -> if you toss a coin, then probability of heads and tails are equally likely right? So, it follows a Uniform Distribution. A Uniform Distribution can be Discrete or Continuous, but for our current topic of discussion lets focus on Continuous Uniform Distribution. 

```python
# Parameters of the distribution
low = 0  # min value of the uniform distribution
high = 50  # max value of the uniform distribution
size = 100000  # number of samples

# Generate samples from the uniform distribution
samples_uniform = np.random.uniform(low=low, high=high, size=size)

# Set the style and color palette
sns.set_style('whitegrid')
sns.set_palette('Dark2')

# Plot the distribution using Seaborn
ax = sns.histplot(samples_uniform, kde=True, alpha=0.7, bins=50)

# Customize the plot
ax.set(xlabel='Value', ylabel='Density', title='Uniform Distribution')
sns.despine(left=True)

# Add vertical lines at the minimum and maximum of the distribution
ax.axvline(x=low, color='red', linestyle='--', linewidth=2, label='Minimum')
ax.axvline(x=high, color='blue', linestyle='--', linewidth=2, label='Maximum')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### See, it works here too, So I think you would agree with me now that it works for all the distributions(untill they follow the rules)

```python
# Generate 1000 sets of 40 random samples from the distribution
means = np.zeros(1000)  # create a NumPy array to store the mean of each set
for i in range(1000):
    sample_set = np.random.choice(samples_uniform, size=40, replace=False)
    mean = np.mean(sample_set)
    means[i] = mean

# Plot a histogram of the means
sns.histplot(means, kde=True)

# Set the x-label, y-label, and title
plt.xlabel('Mean Value')
plt.ylabel('Density')
plt.title('Histogram of Means')

# Show the plot
plt.show()
```
